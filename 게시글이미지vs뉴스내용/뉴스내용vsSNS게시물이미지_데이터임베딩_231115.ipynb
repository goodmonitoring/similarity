```python
from konlpy.tag import Mecab
import numpy as np
from gensim.models import Word2Vec
```

```python
mecab = Mecab()
```

게시물이미지 - 항목 임베딩

```python
df['게시물이미지'] = df['게시물이미지'].str.replace('[^ㄱ-ㅎ ㅏ-ㅣ 가-힝]', '')
```

```python
# 토큰화 함수 생성

def tokenize(doc):
    return [mecab.morphs(doc)]
```

```python
sns_token = []
for i in df['게시물이미지']:
    sns_token.append(tokenize(i))
```

```python
new_df = pd.DataFrame(sns_token)
new_df
```

```python
model = Word2Vec(new_df[0], window=3, workers=1)
```

```python
word_vectors=model.wv
vocabs = word_vectors.index_to_key
word_vectors_list = [word_vectors[v] for v in vocabs]
```

```python
model.wv.vectors.shape
```

```python
def get_document_vectors(document_list):
    document_embedding_list = []

    # 각 문서에 대해서
    for line in document_list:
        doc2vec = None
        count = 0
        for word in line:
            if word in model.wv:
              count += 1
              # 해당 문서에 있는 모든 단어들의 벡터값을 더한다.
              if doc2vec is None:
                doc2vec = model.wv.get_vector(word)
              else:
                doc2vec = doc2vec + model.wv.get_vector(word)

        if doc2vec is not None:
            # 단어 벡터를 모두 더한 벡터의 값을 문서 길이로 나눠준다.
            doc2vec = doc2vec / count
            document_embedding_list.append(doc2vec)

        if doc2vec is None:
            document_embedding_list.append(np.nan)


    # 각 문서에 대한 문서 벡터 리스트를 리턴
    return document_embedding_list
```

```python
document_embedding_list = get_document_vectors(new_df[0])
```

```python
len(document_embedding_list)
```

```python
df['게시물이미지'] = document_embedding_list
```

```python
df = df.dropna()
```

```python
df = df.reset_index(drop = True) # df에 저장시켜줘야 정보가 저장됨.
```

```python
df.info()
```

```python
sns_vector = []

for i in df['게시물이미지']:
  sum = 0

  for j in i:
    sum += j

  sns_vector.append(sum/100)
```

```python
df['게시물이미지'] = sns_vector
```
